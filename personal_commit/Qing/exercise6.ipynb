4. Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time?

```python
text = nltk.corpus.state_union.words()
text
```

```python
fdist = nltk.FreqDist(w.lower() for w in text)
modals = ['men', 'women', 'people']
for m in modals:
     print(m + ':', fdist[m], end=' ')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 5))

cfd = nltk.ConditionalFreqDist(
          (target, fileid[:4])
          for fileid in nltk.corpus.state_union.fileids()
          for w in nltk.corpus.state_union.words(fileid)
          for target in ['men', 'women', 'people']
          if w.lower().startswith(target))
cfd.plot()
```

16. Write a program to generate a table of lexical diversity scores (i.e. token/type ratios), as we saw in 1.1. Include the full set of Brown Corpus genres (nltk.corpus.brown.categories()). Which genre has the lowest diversity (greatest number of tokens per type)? Is this what you would have expected?

```python
def lexical_diversity(my_text_data):
    word_count = len(my_text_data)
    vocab_size = len(set(my_text_data))
    diversity_score = vocab_size / word_count
    return diversity_score
```

```python
nltk.corpus.brown.words(categories="humor")
```

```python
genres = []
diversity = []
for genre in nltk.corpus.brown.categories():
    genres.append(genre)
    lexical_div = lexical_diversity(nltk.corpus.brown.words(categories=genre))
    diversity.append(lexical_div)
```

```python
import numpy as np
import pandas as pd
```

```python
pd.DataFrame(np.matrix([genres, diversity]).T, columns = ["genre","lexical diversity"])
```

* learned has the lowest diversity
