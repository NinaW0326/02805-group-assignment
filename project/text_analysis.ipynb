```python
character_list = list(characters)
char_list = []
for character in character_list:
    char_list.append(character.lower().replace('_',' '))
character_list = char_list
print('done!')
```

```python
# assign the sentences to particular characters
# 1. accoring to punctuations divide book to lists of sentences
import re

sentences_book = re.split("(?<=[.!?])\s+", book.lower()) 
len(sentences_book)
```

```python
# for some important characters
# harry potter
character = 'harry potter '
names = ['harry','potter']
content = ''
path = 'data\{}.txt'.format(character)
for sentence in sentences_book:
    for name in names:
        if name in sentence:
            content += sentence + '\n'
            continue
            
if len(content) > 0: 
    print('rewirting...')
    f = open(path, 'w', encoding ='utf-8') # overwirte
    f.write(content)
    f.close()
print('done!')
```

```python
# sentiment analysis
import os
from textblob import TextBlob

sentiment_dict = {}
filepath = './data/'
files = os.listdir(filepath) # get all the txt names
for file in files:
    path = filepath+file
    f = open(path,'r', encoding = 'utf-8')
    text = f.read()
    blob = TextBlob(text)
    positive_score = blob.sentiment.polarity
    sentiment_dict[file.strip('.txt')] = (positive_score,len(text))
print('done!')
```

```python
sentiment_dict= dict(sorted(sentiment_dict.items(), key=lambda item:item[1][1], reverse = True))
sentiment_dict
```
